```python
import os
from langchain_groq import ChatGroq
from langchain_neo4j import Neo4jGraph, GraphCypherQAChain, Neo4jVector
from langchain_community.embeddings.fastembed import FastEmbedEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_experimental.graph_transformers import LLMGraphTransformer
from langchain_core.documents import Document
```

```python
import os
from langchain_groq import ChatGroq
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_neo4j import Neo4jGraph, GraphCypherQAChain, Neo4jVector
from langchain_experimental.graph_transformers import LLMGraphTransformer
from langchain_core.documents import Document


# Initialize Local Embeddings (Runs on your CPU, No API Key needed)
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

# Initialize LLM (Groq)
llm = ChatGroq(model_name="openai/gpt-oss-120b")

# The Data
text = """
Elon Reeve Musk (born June 28, 1971) is a businessman and investor.
He is the founder of SpaceX and CEO of Tesla.
He dropped out of Stanford and, with his brother Kimbal, co-founded Zip2.
"""
documents = [Document(page_content=text)]

# --- STEP 2: Vector RAG (Using FAISS) ---
# This creates a local FAISS index in memory
vector_db = FAISS.from_documents(documents, embeddings)

def query_vector(query):
    # Searches based on semantic math
    result = vector_db.similarity_search(query, k=1)
    return result[0].page_content

# --- STEP 3: Graph RAG (Using Neo4j) ---
graph = Neo4jGraph()

# Extract relationships from text (Elon -> COFOUNDED -> Zip2)
transformer = LLMGraphTransformer(llm=llm)
graph_docs = transformer.convert_to_graph_documents(documents)
graph.add_graph_documents(graph_docs)

# Chain to turn natural language into Cypher queries
graph_chain = GraphCypherQAChain.from_llm(
    llm=llm, graph=graph, verbose=True, allow_dangerous_requests=True
)

# --- STEP 4: Combined RAG (Neo4j Vector) ---
# Neo4j allows storing both vectors and graph links in one place.
# We use FAISS for standalone vector search, but for 'Combined',
# we use Neo4j's internal vector index + its graph power.
hybrid_db = Neo4jVector.from_documents(
    documents,
    embeddings,
    url=os.environ["NEO4J_URI"],
    username=os.environ["NEO4J_USERNAME"],
    password=os.environ["NEO4J_PASSWORD"]
)

# --- STEP 5: Execution ---
user_query = "Who co-founded Zip2 with Elon Musk?"

print("\n--- 1. VECTOR RAG (FAISS) ---")
# Best for finding the relevant paragraph
print(query_vector(user_query))

print("\n--- 2. GRAPH RAG (NEO4J) ---")
# Best for multi-hop logic (Finding links between people)
print(graph_chain.invoke({"query": user_query})["result"])

print("\n--- 3. COMBINED RAG (NEO4J HYBRID) ---")
# Best for full context + semantic search
print(hybrid_db.similarity_search(user_query)[0].page_content)
```
